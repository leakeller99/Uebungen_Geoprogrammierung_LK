Plotten Sie 1000 Zufallszahlen im Bereich von [-100, -100] bis [100,100] mit verschiedenen	Farben.	Verwenden	Sie	dazu	matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import math
```

```python
# 1000 Zufallszahlen von -100 bis 100
anzahl_punkte = 1000
x = np.random.uniform(-100, 100, anzahl_punkte)
y = np.random.uniform(-100, 100, anzahl_punkte)

#Zufallsfarben fÃ¼r die Punkte
farben = np.random.rand(anzahl_punkte)

#Plot der Zufallszahlen, verschiedene Farben
plt.scatter(x, y, c=farben, marker='o')
plt.title('Zufallszahlen')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def f(x, y):
    return np.exp(-x**2) * np.sin(y)

# Erstelle Arrays fÃ¼r x- und y-Koordinaten
x = np.linspace(-2.5, 2.5, 10000)
y = np.linspace(-2.5, 2.5, 10000)  

# Erstelle ein zweidimensionales Gitter
X, Y = np.meshgrid(x, y)

# Werte die Funktion f(x, y) auf dem Gitter aus
Z = f(X, Y)

# Plotten der Funktion
plt.pcolormesh(X, Y, Z)
plt.colorbar(label='f(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$f(x, y) = \exp(-x^2) \sin(y)$')
plt.show()
```
